```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(4,15,0.3)
```

```python
x
```

```python
plt.plot(x)
plt.show()
```

```python
plt.plot(x)
plt.plot(x**2)
plt.plot(x**3)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,'o--',x**2,'s')
plt.title("BASIC PLOTS")
plt.xlabel('o means circle , s means square')
plt.ylabel('exponential expansion')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,'o--',label='baseline')
plt.plot(x**2,'>:',label='power 2')
plt.title("BASIC PLOTS")
plt.xlabel('o means circle , s means square')
plt.ylabel('exponential expansion')
plt.legend()
plt.show()
```
