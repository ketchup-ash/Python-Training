```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv("data/Pokemon.csv",encoding="latin1",index_col=0)
```

```python
pokemon.head()
```

```python
pokemon.info()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='Attack',y='Defense',
           data=pokemon,hue='Stage',
           fit_reg=False)
plt.show()
```

```python
sns.catplot(x='Defense',y='HP',data=pokemon)
```

```python
plt.
plt.figure(figsize=(11,4))
sns.set_style('whitegrid')
sns.violinplot(x='Type 1',y='Attack',data=pokemon)
plt.show()
```
